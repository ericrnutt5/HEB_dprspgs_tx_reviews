# Workflow for Analyzing Common Words in Negative Store Reviews

### Store: HEB, 598 EHwy, US-290, Dripping Springs, TX 78620

```python
# Import necessary packages
import pandas as pd
from collections import Counter
```

```python
# Import csv as pandas dataframe
heb_dstx_path = '/Users/ericnutt/Desktop/data_analytics_projects/HEB/heb_dstx_reviews.csv'

heb_dstx_df = pd.read_csv(
    heb_dstx_path)

heb_dstx_df.head()
```

#### Pretty sure pd.read_csv parameters can run a lot of the code below

```python
# Convert to datetime
heb_dstx_df['review_datetime_utc'] = pd.to_datetime(heb_dstx_df['review_datetime_utc'])
```

```python
# Remove timestamp from Date and store it in a new column
heb_dstx_df['Date'] = heb_dstx_df['review_datetime_utc'].dt.date
heb_dstx_df.head()
```

```python
# Set the index to the DATE column
heb_dstx_reviews = heb_dstx_df.set_index('Date')
heb_dstx_reviews.head()
```

```python
# Keep only author_title, review_text, review_rating, review_datetime_utc
heb_dstx_reviews_df = heb_dstx_reviews[['author_title','review_text','review_rating']]
heb_dstx_reviews_df
```

```python
# Drop rows with NaN values (not comment left with rating)
heb_dstx_reviews_new = heb_dstx_reviews_df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
heb_dstx_reviews_new.sort_index()
```

```python
# Convert index to datetime
heb_dstx_reviews_new.index = pd.to_datetime(heb_dstx_reviews_new.index)
heb_dstx_reviews_new.index
```

```python
# Parse dates for 2019-2022
heb_dstx_2019_2022 = heb_dstx_reviews_new.sort_index().loc['2019':'2022']
pd.set_option('display.max_colwidth', None)
heb_dstx_2019_2022
```

```python
# Get most common words from review text
Counter(" ".join(heb_dstx_2019_2022["review_text"].str.lower()).split()).most_common(100)
```
